```python
def ImageScarpe(imgnumb):
    sleep(4)
    page_source = driver.page_source
    #print(page_source)
    soup2 = BeautifulSoup(page_source, 'lxml')
    #print(soup)
    type(soup2)

    bs4.BeautifulSoup

    title = soup2.title 


    # get the image source
    #itemprop="image"
    
    
    anchor_tag = soup2.find("a", {"class":"college_logo_wrap"})
    logoimg_tag=anchor_tag.find("img")
 
    div_tag =  soup2.find("div", {"class":"college_top_wrapper"}) 
    Bannerimg_tag = div_tag.find_all("img")[0]
    #img-responsive bg-logo lazyloaded
   
    # Workbook() takes one, non-optional, argument 
    # which is the filename that we want to create. 
    

    # The workbook object is then used to add new 
    # worksheet via the add_worksheet() method. 
    
    #src = img.get_attribute('src')
    
    #print(image["src"])

    # This is the image url.
    image_url = logoimg_tag["src"]
    Banner_url=Bannerimg_tag["src"]
    
    #print(i,image_url)
    # Open the url image, set stream to True, this will return the stream content.
    resp = requests.get(image_url, stream=True)
    resp1= requests.get(Banner_url, stream=True)
    
    # Open a local file with wb ( write binary ) permission.
    #Logo
    local_file = open("D:/CollegeDunia/Imagescraper/INDIAN/LOGOS//logo"+ str(imgnumb) +".jpg", 'wb')
    #Banner
    local_file1 = open("D:/CollegeDunia/Imagescraper/INDIAN/BANNERS//Banner"+ str(imgnumb) +".jpg", 'wb')

    # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
    resp.raw.decode_content = True
    resp1.raw.decode_content = True
    #print(local_file)

    # Copy the response stream raw data to local image file.
    shutil.copyfileobj(resp.raw, local_file)
    shutil.copyfileobj(resp1.raw, local_file1)
    del resp
```

```python
def DatatoExcelcreation(startlink,endlink):

    from openpyxl import Workbook
    import openpyxl
    #create Workbook object
    wb=Workbook()
    #set file path
    #filepath="/home/ubuntu/demo.xlsx"
    # select demo.xlsx
    data=[("College Name:","LOGo","Established Year","Rating","Web LInk","Phone Number","City&State","Country","Full Address","overview","College_Highlights","college_Scholarships","college_Admission","college_Placement","Course Name","Course Type:","Eligibility","streams","Duration","Mode","Total Fee")]
    sheet=wb.active
    #data=[("id","NAME","EDuCATION")]
    for row in data:
        sheet.append(row)

    for i in range(0,len(collegname_list)):

        data=[(collegname_list[i],logo_list[i],ESTDYears_List[i],Rating_list[i],weblink_list[i],phone_list[i],CItyState_List[i],country_list[i],address_list[i],college_overview[i],college_Highlights[i],college_Scholarships[i],college_Admission[i],college_Placement[i],course_list[i],coursetype_list[i],elgbtly_list[i],stream_list[i],courseduration_list[i],coursemode_list[i],TotalFee_list[i])]
        for row in data:
            sheet.append(row)
    # append all rows
    k=startlink
    for i in range(0,len(Imageindex_list)-1):
        #print(k)
        print(Imageindex_list[i])
        try:
            img = openpyxl.drawing.image.Image("D:/CollegeDunia/Imagescraper/INDIAN/LOGOS//logo"+ str(k+1) +".jpg")
            #print(img)
            img.anchor = Imageindex_list[i]
            sheet.add_image(img)
            k=k+1
            print("Inserting done from TRY ")
        except:
            print(str(k+1)+" no image found error")
            k=k+1

    # save file
    wb.save(r"D:\CollegeDunia\DataBase\arts-colleges("+str(startlink)+"-"+str(endlink)+").xlsx")
    
    print("**************************")
    print("excel creation done sucess")
    print("**************************")
```
